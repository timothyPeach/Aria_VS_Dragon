# Planning Game Structure

### Aria vs The Dragon

#### Aria's stats:

HP = 255 #health

ATK = 150 +- 25 #physical attack (sword)

DEF = 100  #physical defense (armor)

SPATK = 255 #magic attack

SPDEF = 200 #magic defense

#### Aria's moves:

##### physical moves: #sword

Attack (deals some amount of physical damage to target entity + random variation)



##### status and healing spells

Heal (heals target entity some amount of health back + random variation)

Shell (improves magic defense of target entity by a fixed amount for next 8 moves)

Protect (improves physical defense of target entity by a fixed amount for next 8 moves)

Reflect (gives target entity a reflective shield for next 8 moves. Any spell that hits this reflective shield will not hit the entity who has the shield, but instead will reflect back and hit the other entity! This is true even if the other entity has reflect on as well.)

note: if you use a spell like shell twice on the same entity, it won't stack, it'll just reset the amount of moves that it will remain active back to 8.



##### damage spells

Fire (deals some amount of magic damage to target entity  + no random variation)

Ice (deals some amount of magic damage to target entity + medium random variation)

Lightning (deals some amount of magic damage to target entity  + high random variation)


#### Dragon's stats:

HP = 1300 #health

ATK = 200 +- 25 #physical attack (claws)

DEF = 100  #physical defense (scales)

SPATK = 255 #magic attack

SPDEF = 100 #magic defense

##### The dragon has all the same moves Aria has, but he only uses Heal on himself and Attack and Ice against opponents.  The dragon could use the other moves, but he doesn't because thinking too hard makes him angry.  

##### Also, every 20 moves, the dragon's ATK and SPATK increase by 20, also because of anger.

#### Lastly, the Dragon has a number of insults at his disposal that he uses in order to demoralize his opponents.  Very important, believe me.

```python
#Physical Damage Area
def attack(playerString,vals):
    plusMinusVariance = random.randint(-25,25)
    return vals[playerString]["ATK"] + plusMinusVariance

#Magic Damage Area

def damageSpell(playerString,spellString,vals):
    if spellString == "f":
        varianceInt = 0
    elif spellString == "i":
        varianceInt = 50
    elif spellString == "l":
        varianceInt = 100
        
    plusMinusVariance = random.randint(-varianceInt,varianceInt)
    return vals[playerString]["SPATK"] + plusMinusVariance

#Healing Area

def heal(playerString,vals):
    return -vals[playerString]["SPATK"]

#Taking Damage Area
def takeDamage(playerString,damage,damageType,vals):
    if damage < 0:
        return damage
    if damageType == 'physical':
        damageString = "DEF"
    else:
        damageString = "SPDEF" 
    out = damage - vals[playerString][damageString]
    return out
```

```python
def getTrainingData(gamesToPlay,model=None,keepDataMultiplier=0.2,randomnessPercent=10):
    '''# for gamesToPlay choose number divisible by 5'''

    rowsKept = []
    rows = []
    minDragonHPs = []

    for i in range(gamesToPlay):
        rows.append(gameForNN(model,randomnessPercent))

    rows = sorted(rows)
    
    forAverage = []
    
    for i in range(len(rows)):
        forAverage.append(rows[i][0])
    print("average lowest Dragon HP: ",np.mean(forAverage))
    print("record for this iteration: ",min(forAverage))
    
    for i in range(len(rows)):
        rows[i].remove(rows[i][0])

    rowsKept = rows[:][:int(gamesToPlay*keepDataMultiplier)]
    
    return rowsKept
```

```python
def convertToDataFrame(rowsKept):
    
    df = pd.DataFrame()
    
    rowDict = {}
    for i in range(9):
        rowDict[i] = []
    for i in range(9):
        for x in range(len(rowsKept)):
            rowDict[i] += rowsKept[x][i]
        
    df["ariaShell"] = rowDict[0]
    df["ariaProtect"] = rowDict[1]
    df["ariaReflect"] = rowDict[2]
    df["dragonShell"] = rowDict[3]
    df["dragonProtect"] = rowDict[4]
    df["dragonReflect"] = rowDict[5]
    df["ariaHP"] = rowDict[6]
    df["dragonHP"] = rowDict[7]
    df["move"] = rowDict[8]
    
    return df
```

```python
def trainForest(df):
    X = df.drop("move",axis=1)
    y = df["move"]
    rfc = RandomForestClassifier(n_estimators=20)
    rfc.fit(X,y)
    #print(rfc.predict(X.values[0].reshape(1,-1)))
    #The output of the prediction is a list containing one element which is the correct move,
    #so when we use it in the next iteration, we need to make sure to take the 0th index of the prediction.
    return rfc
```

```python
def evolvingForests(iterations,gamesPerIteration=10000,keepDataMultiplier=0.2,randomnessPercent=10,initialModel=None):
    '''iterations should be a minimum of 1.'''
    if initialModel == None:
        daModel = trainForest(convertToDataFrame(getTrainingData(gamesPerIteration)))
        iterations -= 1
    else:
        daModel = initialModel
    while iterations != 0:
        daModel = trainForest(convertToDataFrame(getTrainingData(gamesPerIteration,model=daModel,randomnessPercent=randomnessPercent)))
        iterations -= 1
        
    return daModel
```

```python
def evaluateCurrentModel(model,gamesToPlay):
    results = 0
    for _ in range(int(gamesToPlay)):
        tempResult = gameForNN(model,exportResult=True,randomness=False)
        results += tempResult

    winRate = results/float(gamesToPlay)

    return winRate
```

```python
def evaluateTotalRandomness(gamesToPlay):
    results = 0
    for _ in range(int(gamesToPlay)):
        tempResult = gameForNN(exportResult=True)
        results += tempResult

    winRate = results/float(gamesToPlay)

    return winRate
```

# You know for now I'm just going to use random forests. I'm actually pretty interested to see how they will perform.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score

rfc = RandomForestClassifier(n_estimators=20)

X = keepers.drop("move",axis=1)
y = keepers["move"]

kf_shuffle = KFold(n_splits=10,shuffle=True,random_state=42)
scores = cross_val_score(rfc, X, y, cv=kf_shuffle)

#print(scores)
print(np.mean(scores))
print(np.std(scores))
print("baseline",max(keepers["move"].value_counts())/float(len(keepers)))
```
