# Planning Game Structure

### Aria vs The Dragon

#### Aria's stats:

HP = 255 #health

ATK = 150 +- 25 #physical attack (sword)

DEF = 100  #physical defense (armor)

SPATK = 255 #magic attack

SPDEF = 200 #magic defense

#### Aria's moves:

##### physical moves: #sword

Attack (deals some amount of physical damage to target entity + random variation)



##### status and healing spells

Heal (heals target entity some amount of health back + random variation)

Shell (improves magic defense of target entity by a fixed amount for next 8 moves)

Protect (improves physical defense of target entity by a fixed amount for next 8 moves)

Reflect (gives target entity a reflective shield for next 8 moves. Any spell that hits this reflective shield will not hit the entity who has the shield, but instead will reflect back and hit the other entity! This is true even if the other entity has reflect on as well.)

note: if you use a spell like shell twice on the same entity, it won't stack, it'll just reset the amount of moves that it will remain active back to 8.



##### damage spells

Fire (deals some amount of magic damage to target entity  + no random variation)

Ice (deals some amount of magic damage to target entity + medium random variation)

Lightning (deals some amount of magic damage to target entity  + high random variation)


#### Dragon's stats:

HP = 1300 #health

ATK = 200 +- 25 #physical attack (claws)

DEF = 100  #physical defense (scales)

SPATK = 255 #magic attack

SPDEF = 100 #magic defense

##### The dragon has all the same moves Aria has, but he only uses Heal on himself and Attack and Ice against opponents.  The dragon could use the other moves, but he doesn't because thinking too hard makes him angry.  

##### Also, every 20 moves, the dragon's ATK and SPATK increase by 20, also because of anger.

#### Lastly, the Dragon has a number of insults at his disposal that he uses in order to demoralize his opponents.  Very important, believe me.

```python
#Physical Damage Area
def attack(playerString,vals):
    plusMinusVariance = random.randint(-25,25)
    return vals[playerString]["ATK"] + plusMinusVariance

#Magic Damage Area

def damageSpell(playerString,spellString,vals):
    if spellString == "f":
        varianceInt = 0
    elif spellString == "i":
        varianceInt = 50
    elif spellString == "l":
        varianceInt = 100
        
    plusMinusVariance = random.randint(-varianceInt,varianceInt)
    return vals[playerString]["SPATK"] + plusMinusVariance

#Healing Area

def heal(playerString,vals):
    return -vals[playerString]["SPATK"]

#Taking Damage Area
def takeDamage(playerString,damage,damageType,vals):
    if damage < 0:
        return damage
    if damageType == 'physical':
        damageString = "DEF"
    else:
        damageString = "SPDEF" 
    out = damage - vals[playerString][damageString]
    return out
```
